```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
url=f"https://gameloot.in/product-category/graphics-card/"
```

```python
page=requests.get(url).text
```

```python
doc=BeautifulSoup(page,"html.parser")
print(doc.prettify)
```

```python
page_text = doc.find(class_="page-numbers")
print(page_text)
```

```python
page_numbers = []
page_number_elements = doc.find_all("a", class_="page-numbers")

for element in page_number_elements:
    page_number = element.get_text()
    if page_number.isdigit():
        page_numbers.append(int(page_number))
print(page_numbers)
```

```python
search_term=gpu=input("what type of GPU do u need: ")
```

## Approach 1 (Using Python Dictionary)

```python
items_found={}
for page in range(1, max(page_numbers)+1):
    url=f"https://gameloot.in/product-category/graphics-card/page/{page}/"
    page=requests.get(url).text
    doc=BeautifulSoup(page,"html.parser")
    div=doc.find(class_="product_details")
    items=div.find_all(string=re.compile(search_term))
    
    
    for item in items:
        #extracting link
        parent = item.find_parent(class_='product_item_link')
        link=None
        if parent:
            link = parent['href']
        
        suburl=link
        a=requests.get(suburl).text
        b=BeautifulSoup(a,"html.parser")
        
        
        #extracting name
        name_xtra=b.find(class_="product_title entry-title")
        name = name_xtra.get_text(strip=True)
        
        
        #extracting cost
        cost_xtra=b.find(class_="pricebox")
        cost = cost_xtra.get_text(strip=True)
        cost_parts = cost.split("Rs.")
        if len(cost_parts) >= 3:
            original_price = cost_parts[1]
            selling_price = cost_parts[2]
        formatted_cost = f"Original Price: Rs.\033[1;32m{original_price}\033[0m    Selling Price: Rs.\033[1;31m{selling_price}\033[0m"

        items_found[name]={"price":formatted_cost,"link":link}
#         sorted_items=sorted(items_found.items())
        #we can use lambda function to sort by price if we take only integer cost in price section 
        
        for item in items_found.items():
            print(f"\033[1;35m{item[0]}\033[0m")
            print(item[1]['price'])
            print(item[1]['link'])
            print()
```
